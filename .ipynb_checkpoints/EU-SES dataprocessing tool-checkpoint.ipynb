# Analysis of the effects of scale on different variables used in the EU-SES dataprocessing tool

## Create three datasets with different using different 
### Dissagregation and aggregation variables

This jupyter inroduces the various functions that make up the EU-SES data processing tool. Specifically this notebook will analyse the effect of using different proxy variables to aggregate and dissagregate spatial variables that are used to build energy system models. The data processing tool builds two datasets when building an energy system model. The areas dataset and the regions dataset. The power demand profile can either be dissagregated using population or gdp. Subsequently, the effect of selecting different varibales to aggregate capacity factors are compared. Two different set of variables are used to aggregate the areas dataset variables to the regions dataset. Hence, three different euses data collections are built with varying sets of variables used to dissagregate power data and subsequently aggregate capacity factor data.

| Dataset collection no. | Dataset collection name| Power disaggregation variable | Aggregation variables |
| ---         |    ----     |             ---               |  ---                  |
| 1           | euses_n1    | Population                    | Set 1                 |
| 2           | euses_n2    | GDP                           | Set 2                 |
| 3           | euses_n3    | Population                    | Set 2                 |

Sets of variables used to aggregate the areas dataset to regions dataset. 

| Variables                    | Set 1        |          Set 2          | 
| ---                         |       ----   |        ---              | 
| Onshore wind CF             | Land Area    | Eligible onshore area  | 
| Solar PV CF                 | Land Area    | Eligible PV area        | 
| Air-sourced heat pump COP   | Land Area    |   Population            | 

A euses data collecion can be build from scrach by using the euses 'build_dataset' function  
example = euses.build_dataset(countries,year)

Alternatively, a pre-build euses collection can be imported using the 'import_dataset' function.
example = euses.import_dataset('eu_2011.nc')

In this jupyter notebook a pre-build euses data collection containing 28 European countries called 'eu_2011.nc' is used. The 'eu_2011.nc' data collection is built using 2011 as the reference year.

### Analyse effects of variation in power demand distributions
A geopandas dataframe of nuts 2 areas is built and the annual power demand of NUTS 2 areas from euses_n1 and euses_n2 named 'power_pop' and 'power_gpd' respectively is added. The ratio and the difference between the two power demand columns are given in 'pow_ratio' and 'pow_diff' respectively. The differnce in annual power demand in relation to the respective national power demand of the NUTS 2 area is calculated and inserted in the 'pow_diff_nat' column. 

```python
ds = euses_n1.ds
nuts2_gdf = gpd.GeoDataFrame(index=ds.coords['nuts_2'],
                            geometry= ds['geometry'].values,
                            data = ds['country_code'].values,
                            columns = ['country_code'],
                            crs='epsg:3035')
    
nuts2_gdf['area'] = nuts2_gdf.geometry.area * 1e-9
nuts2_gdf['pop_density'] = ds.population.values/nuts2_gdf.geometry.area

nuts2_gdf['power_pop'] = euses_n1.ds_regions['power'].sum(dim='time')
nuts2_gdf['power_gdp'] = euses_n2.ds_regions['power'].sum(dim='time')

nuts2_gdf['country_code'] = ds['country_code']
#nuts2_gdf['pow_ratio'] = (nuts2_gdf['power_pop']-nuts2_gdf['power_gdp'])/nuts2_gdf['power_gdp']
#nuts2_gdf['pow_ratio'] = (nuts2_gdf['power_gdp']-nuts2_gdf['power_pop'])/nuts2_gdf['power_pop']

nuts2_gdf['pow_ratio'] = nuts2_gdf['power_gdp']/nuts2_gdf['power_pop']




#nuts2_gdf['pow_ratio'] = (nuts2_gdf['power_gdp'])/nuts2_gdf['power_pop']
nuts2_gdf['pow_diff'] = (nuts2_gdf['power_gdp']-nuts2_gdf['power_pop'])*1e-6
```

```python
# Calculate the NUTS 2 areas relative difference in power demand in relation to their respective national demand. 
# Use box plot o display result.
# With the exeption of Italy, Austria and Norway all countries experience a reduction in the median power demand 
for i,rows in nuts2_gdf.iterrows():
    c_pow = nuts2_gdf.loc[nuts2_gdf.country_code==rows.country_code]['power_gdp'].sum()*1e-6
    nuts2_gdf.loc[i,'pow_diff_nat'] = nuts2_gdf.loc[i,'pow_diff']/c_pow*100
```

```python
def boxplot_sorted(df, by, column):
    fig, axs = plt.subplots(figsize=(8,5))
    df2 = pd.DataFrame({col:vals[column] for col, vals in df.groupby(by)})
    df2 = df2.drop(['EE','LU','LT','LV'],axis=1)
    meds = df2.median().sort_values()
    df2[meds.index].boxplot(ax=axs, rot=90)
    axs.set_xlabel('Countries')
    axs.set_ylabel('Difference in annual power demand\nrelative to national power demand (%)')


boxplot_sorted(nuts2_gdf, by=["country_code"], column="pow_diff_nat")
```

Box plot shows that as the medians increase the maximum outlier values also tend to decrease. Most countries with negative medians have outliers that are more than twice the value of the upper fence. These countries outliers are greater than 0.05.

### Analyse effects of aggregation on capacity factors 

```python
euses_n1.create_regions('nuts0', aggr_var = sets['set1'], disaggr_var = {'power':'population'})
euses_n2.create_regions('nuts0', aggr_var = sets['set2'], disaggr_var = {'power':'gdp'})
euses_n3.create_regions('nuts0', aggr_var = sets['set2'], disaggr_var = {'power':'population'})
```

```python
ds = euses_n1.ds
nuts2_gdf = gpd.GeoDataFrame(index=ds.coords['nuts_2'],
                            geometry= ds['geometry'].values,
                            data = ds['country_code'].values,
                            columns = ['country_code'],
                            crs='epsg:3035')
    
nuts2_gdf['area'] = nuts2_gdf.geometry.area * 1e-9
```

```python
# Calculate ratio between areas mean and respective national mean of capacity factors
for euses,sets in {euses_n1:'_set1', euses_n2:'_set2'}.items():
    for cf in ['wind_cf','pv_cf','cop_air']:
        for i, rows in nuts2_gdf.iterrows():
            id = i
            nuts2_gdf[cf] = euses.ds[cf].mean(dim='time')
            areas = euses_n1.ds_regions.country_code.where(euses_n1.ds_regions.country_code==rows.country_code,drop=True).coords['regions'].values
            nuts2_gdf.loc[id,cf+sets+'_dev'] = nuts2_gdf.loc[id,cf]/euses.ds_regions[cf].loc[areas].mean().values

            
```

#### Visualise difference between national mean and NUTS 2 area mean 

```python
def plot_mean_difference(cf_var,cmap,vcenter=1):
#    min_cf =  round(nuts2_gdf[cf_var].min()*100,-1)/100
#    max_cf = 1 + (1-min_cf)
    min_cf = round(nuts2_gdf[cf_var].min(),2)
    max_cf = round(nuts2_gdf[cf_var].max(),2)    
    fig, ax = plt.subplots(figsize=(5, 5))
    norm=mpl.colors.TwoSlopeNorm(vmin=min_cf,vcenter=vcenter, vmax=max_cf)
    nuts2_gdf.plot(ax=ax,column=cf_var,
                   cmap=cmap,
                   norm=norm,
                   edgecolor="#c2c2c2")
    ax.set_axis_off()
    fig.colorbar(mpl.cm.ScalarMappable(norm=norm, cmap=cmap), shrink=0.7, label='National mean to NUTS 2 area mean ratio')
```

```python
plot_mean_difference('wind_cf_set1_dev','RdBu_r'),plot_mean_difference('wind_cf_set2_dev','RdBu_r')
```

```python
plot_mean_difference('pv_cf_set1_dev','PuOr_r'),plot_mean_difference('pv_cf_set2_dev','PuOr_r')
```

```python
plot_mean_difference('cop_air_set1_dev','RdGy_r'),plot_mean_difference('cop_air_set2_dev','RdGy_r')
```

#### A T-Test is conducted on the capacity factors to test the significance difference between national mean and sub national areas.
The T-Tests are conducted for solar PV, onshore wind and ASHP capacity factors to compare the two sets of variables used to aggregate the capacity factors.  


```python
# A t-test to measure statistical difference of capacity factors calculated 
# using set 1 or using set 2 variables
```

```python
national_df = pd.DataFrame(index=euses_n1.ds_regions.country_code)
national_df['regions'] = euses_n1.ds_regions.coords['regions']
national_df['country_code'] = national_df.index
national_df = national_df.drop(['EE','LU','LT','LV'],axis=0)
for id,rows in national_df.iterrows():
    reg = rows.regions
    for cf_var in ['wind_cf','pv_cf','cop_air']:
        reg_s1_var = euses_n1.ds_regions[cf_var].sel(regions=reg).transpose()
        reg_s2_var = euses_n2.ds_regions[cf_var].sel(regions=reg).transpose()
        national_df.loc[id,cf_var+'_pvalue'] = ttest_ind(reg_s1_var, reg_s2_var)[1]
        national_df.loc[id,cf_var+'_ratio'] = (reg_s1_var/reg_s2_var).mean()
```

```python
for cf_var in ['wind_cf','pv_cf','cop_air']: 
    fig, ax = plt.subplots(figsize=(8,2))
    sorted_df = national_df.sort_values(by=[cf_var+'_ratio'])
    sorted_df.plot.scatter(ax=ax, x='country_code',y=cf_var+'_ratio')
    filtered_df = sorted_df.query('{}_pvalue<0.05'.format(cf_var))
    filtered_df.plot.scatter(ax=ax, x='country_code',y=cf_var+'_ratio',color='r')
    ax.set_xlabel('Countries')
    ax.set_ylabel('Difference between weighted national mean\ncapacity values')
```
