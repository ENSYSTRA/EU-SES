# Building three case studies using the EU-SES modelling tool
This jupyter notebook demonstrates the versitality of the EU-SES modelling tool by building three energy models with varying spatial context. The spatial characteristics that are varied are the spatail classification method and the spatial scope.
The characteristics of the three case studies are defined in the table below.

| Model name |                                                  Countries                                                  | Classification method | Number of regions | Number of NUTS 2 areas |
|:----------:|:-----------------------------------------------------------------------------------------------------------:|:---------------------:|:-----------------:|:----------------------:|
| GER NUTS0  | Germany, Norway, Denmark, Poland, France, Netherlands, Belgium, Austria, Switzerland and the Czech Republic | National boundaries   | 10                |  134                      |
| GER MAX-P  | Germany                                                                                                     | Max-p regions         | 9                 | 38                     |
| GER NUTS1  | Germany                                                                                                     | State boundaries      | 16                | 38                     |


The demonstration is structured in the following sections:
1. Build or import Areas dataset 
2. Build case studies
3. Import case studies model result
4. Plot results

In steps 1 and 2 the models can be built from scratch by first building the Areas dataset and subsequetly building and solving the models. As the models have been pre-built and presolved, step 1 and 2 can be skipped and the models can be imported instead in step 3. In step 4 the results of the models are visualised. 

For all steps the neccesary packages for this notebook need to be imported.

```python
# Import the neccesary pyton packages and the euses library
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
import calliope
import numpy as np
import warnings
import seaborn as sns

import euses
warnings.filterwarnings("ignore")
```

## 1. Build or import Areas dataset 
The next step is to obtain an Areas dataset. The areas dataset can either be built (Option 1) or imported from a pre-buit dataset (Option 2). The reference year and the countries of interest are defined first.

```python
reference_year = 2011
countries = ['Germany','Norway','Denmark','Poland','France','Netherlands',
                                 'Belgium','Austria','Switzerland','Czech Rep.']
```

## Build the Areas dataset  (Option 1)
This step takes approximately 25 minutes as data variables needed to build the Areas dataset are downloaded and processed.The Areas dataset is build using the euses.build_dataset() function for a selection of European counries that will be used in the three models. In this case study the countries that will be used are Germany, Norway, Denmark, Poland, France, Netherlands, Belgium, Austria, Switzerland and the Czech Republic. A reference year is also given to the function to select variables, where available, from a specific year of interest. The reference year used in all case studies is 2011. 

```python
# case study dataset
cs_ds = euses.build_dataset(countries,reference_year)
```

## Import the Areas dataset (Option 2)


```python
# eu wide dataset imported 
# countries of interest filtered in cs_ds
eu_ds = euses.import_dataset('eu_2011.nc')
cs_ds = eu_ds.filter_countries(countries)
```

# 2. Build case studies
In all case studies the optimised energy system models is constrained by the cummulaive amount of CO2 that can be emitted from the power plants. The energy models CO2 emmissions are limited to 20% of the 1990 emission level.  

## Filtering power plants of interest
Taking into consideration the before mentioned constraint of the scenarios, the types of technologies that are included in the model is limited to only existing renewable energy power plants and plants that operate on natural gas.

```python
fuels_considered = ['Biomass and biogas','Natural gas','Solar','Wind']
fuels_removed = np.setdiff1d(cs_ds.ds.coords['fuel'].values,fuels_considered)
cs_ds.ds = cs_ds.ds.drop(fuels_removed,dim='fuel')
tech_not_grouped = ['Solar','Wind','Wind Offshore']
tech_list = np.setdiff1d(cs_ds.ds.coords['tech'].values,tech_not_grouped)
# group natural gas and biomass/biogas based power_plants in single group
sum_var = cs_ds.ds['power_plants'].loc[:,tech_list,:].sum(axis=1)
tech_list = np.setdiff1d(tech_list,['Combined cycle'])
cs_ds.ds['power_plants'].loc[:,['Combined cycle']] = [[g] for g in sum_var.values]
cs_ds.ds = cs_ds.ds.drop(tech_list,dim='tech')
```

## 2.1 Build and optimise first case study model - GER NUTS0

```python
cs_ds.create_regions('nuts0')
cs_ds.create_calliope_model(op_mode='plan',sectors=['power','heat'],co2_cap_factor=0.2, national=True)
ger_nuts0 = calliope.Model('calliope_model/model.yaml',scenario='time_3H',override_dict={'run.solver': 'cbc'})
ger_nuts0_gpd = gpd.GeoDataFrame(cs_ds.ds_regions['regions'].values,
    columns=['id'],geometry=cs_ds.ds_regions['geometry'].values)
ger_nuts0.run()
ger_nuts0.to_netcdf('calliope_model/results/cs_ger_nuts0.nc')
```

## 2.2 Build and optimise second case study model - GER MAX-P

```python
cs_ds = cs_ds.filter_countries(['Germany'])
cs_ds.create_regions('max_p_regions',3.8)
cs_ds.create_calliope_model(op_mode='plan',sectors=['power','heat'],co2_cap_factor=0.2, national=False)
ger_max_p = calliope.Model('calliope_model/model.yaml',scenario='time_3H',override_dict={'run.solver': 'cbc'})
ger_nuts0_gpd = gpd.GeoDataFrame(cs_ds.ds_regions['regions'].values,
    columns=['id'],geometry=cs_ds.ds_regions['geometry'].values)
ger_max_p.run()
ger_max_p.to_netcdf('calliope_model/results/cs_ger_max_p.nc')
```

## 2.3 Build and optimise third case study model - GER NUTS1

```python
cs_ds = cs_ds.filter_countries(['Germany'])
cs_ds.create_regions('nuts1')
cs_ds.create_calliope_model(op_mode='plan',sectors=['power','heat'],co2_cap_factor=0.2, national=False)
ger_nuts1 = calliope.Model('calliope_model/model.yaml',scenario='time_3H',override_dict={'run.solver': 'cbc'})
ger_nuts1_gpd = gpd.GeoDataFrame(cs_ds.ds_regions['regions'].values,
    columns=['id'],geometry=cs_ds.ds_regions['geometry'].values)
ger_nuts1.run()
ger_nuts1.to_netcdf('calliope_model/results/cs_ger_nuts1.nc')
```

## 3. Import case studies model results

```python
# Import solved opimisation models
ger_nuts0 = calliope.read_netcdf('calliope_model/results/cs_ger_nuts0.nc')
ger_nuts1 = calliope.read_netcdf('calliope_model/results/cs_ger_nuts1.nc')
ger_max_p = calliope.read_netcdf('calliope_model/results/cs_ger_max_p.nc')
```

```python
# building GER NUTS0 regions
eu_ds = euses.import_dataset('eu_2011.nc')

countries = ['Germany','Norway','Denmark','Poland','France','Netherlands',
                                 'Belgium','Austria','Switzerland','Czech Rep.']
cs_ds = eu_ds.filter_countries(countries)
cs_ds.create_regions('nuts0')
ger_nuts0_gpd = gpd.GeoDataFrame(cs_ds.ds_regions['regions'].values,
    columns=['id'],geometry=cs_ds.ds_regions['geometry'].values)

# building GER MAXP regions
cs_ds = cs_ds.filter_countries(['Germany'])
cs_ds.create_regions('max_p_regions',3.8)
ger_max_p_gpd = gpd.GeoDataFrame(cs_ds.ds_regions['regions'].values,
    columns=['id'],geometry=cs_ds.ds_regions['geometry'].values)

# building GER NUTS0 regions
cs_ds.create_regions('nuts1')
ger_nuts1_gpd = gpd.GeoDataFrame(cs_ds.ds_regions['regions'].values,
    columns=['id'],geometry=cs_ds.ds_regions['geometry'].values)
```

## 4. Plot results

```python
# Calculate area usage by model and technology

area_usage = pd.DataFrame(index=['GER NUTS0', 'GER NUTS1', 'GER MAX-P'])
results_dic={'GER NUTS0':ger_nuts0, 'GER NUTS1':ger_nuts1, 'GER MAX-P':ger_max_p}

for name, m in results_dic.items():
    for tech in ['wind','solar','wind_offshore']:
        filter = {'techs':tech}
        if name == 'GER NUTS0':
            filter = {'techs':tech,'locs':'DE'}
        area_max = m.get_formatted_array('resource_area_max').loc[filter].sum()
        area_used = m.get_formatted_array('resource_area').loc[filter].sum()
        area_usage.loc[name,tech] = area_used/area_max*100

# Calculate curtailment by model and technology
curtailment = pd.DataFrame(index=['GER NUTS0', 'GER NUTS1', 'GER MAX-P'])

for name, m in results_dic.items():
    for tech in ['wind','solar','wind_offshore']:
        filter = {'techs':tech}
        if name == 'GER NUTS0':
            filter = {'techs':tech,'locs':'DE'}
        flh = m.get_formatted_array('carrier_prod').loc[dict(filter,**{'carriers':'power'})].sum()/m.get_formatted_array('energy_cap').loc[filter].sum()
        flh_resource = m.get_formatted_array('resource').loc[filter].sum()
        curtailment.loc[name,tech] = (1-flh/flh_resource).values*100


techs= ['solar','wind', 'wind_offshore','combined_cycle', 'heat_pump_air', 'hdam',
        'hphs', 'hror', 'hydrogen']

ger_max_p.inputs['colors'].loc['dc_transmission'] = '#D2BB5F'
colors = ger_max_p.inputs['colors'].loc[techs+['ac_transmission','dc_transmission']].values

# load installed capaciy of technologies from model results into pandas dataframe
df_IC = pd.DataFrame()
fig_dic = {ger_nuts0:'GER_NUTS0',ger_nuts1:'GER_NUTS1',ger_max_p:'GER_MAXP'}

for model, name in fig_dic.items():
    if name == 'GER_NUTS0':
        installed_capacity = model.get_formatted_array('energy_cap').loc['DE',:]
    else:
        installed_capacity = model.get_formatted_array('energy_cap').sum(axis=0)
    df_IC[name] = installed_capacity.loc[techs].to_pandas()
    for transmission in ['ac_transmission','dc_transmission']:
        df_IC.loc[transmission,name] = installed_capacity.loc[installed_capacity.coords['techs'].str.contains(transmission)].sum()
df_IC=df_IC/1e6 # convert to TW from GW

def plot_graphs(w,l):
    fig3 = plt.figure(figsize=(w,l),frameon=False)
    gs = fig3.add_gridspec(4, 3)

    # fig (a)
    cmap = sns.color_palette("Spectral", as_cmap=True)
    ax_a1 = fig3.add_subplot(gs[0, 0])
    ger_nuts0_gpd.plot(ax=ax_a1,column='id',cmap=cmap, edgecolor='grey')
    ax_a1.set_title(label='GER NUTS0')
    ax_a2 = fig3.add_subplot(gs[0, 1])
    ger_nuts1_gpd.plot(ax=ax_a2,column='id',cmap=cmap,edgecolor='grey')
    ax_a2.set_title(label='GER NUTS1')
    ax_a3 = fig3.add_subplot(gs[0, 2])
    ger_max_p_gpd.plot(ax=ax_a3,column='id',cmap=cmap,edgecolor='grey')
    ax_a3.set_title(label='GER MAX-P')

    [axs.set_axis_off() for axs in [ax_a1,ax_a2, ax_a3]]

    # fig (b)
    ax_b = fig3.add_subplot(gs[1, :])
    techs_name= ['Solar','Onshore wind', 'Offshore wind','Cogeneration',
     'Air-sourced heat pump', 'Reservoir-based hydropower',
     'Pumped-storage', 'Run-of-river hydropower', 'Hydrogen','AC Transmission',
     'DC Transmission']

    df_IC.transpose().plot.bar(ax=ax_b,stacked=True, color=colors)
    ax_b.legend(techs_name, loc='lower left', bbox_to_anchor=(1, -0.1),ncol=1)
    ax_b.set_ylabel('Installed \nCapacity (TW)')

    ax_b.set_xticklabels(['GER NUTS0', 'GER NUTS1', 'GER MAX-P'], rotation=0)
    ax_b.tick_params(length=0)

    for side in ['top','right']:
        ax_b.spines[side].set_visible(False)

    # fig (c)
    ax_c = fig3.add_subplot(gs[2, :])
    colors_vre = ger_max_p.inputs['colors'].loc[curtailment.columns].values
    curtailment.plot.bar(ax=ax_c,color=colors_vre)

    ax_c.legend(curtailment.columns, loc='lower left', bbox_to_anchor=(1, 0),ncol=1)
    ax_c.set_ylabel('Percentage \ncurtailment')

    ax_c.set_xticklabels(['GER NUTS0', 'GER NUTS1', 'GER MAX-P'], rotation=0)
    ax_c.tick_params(length=0)
    ax_c.legend(['Onshore wind','Solar','Offshore wind'], loc='lower left', bbox_to_anchor=(1, 0.1),ncol=1)

    for side in ['top','right']:
        ax_c.spines[side].set_visible(False)

    # fig (d)

    ax_d = fig3.add_subplot(gs[3, :])
    colors_vre = ger_max_p.inputs['colors'].loc[curtailment.columns].values
    area_usage.plot.bar(ax=ax_d,color=colors_vre)

    ax_d.legend(curtailment.columns, loc='lower left', bbox_to_anchor=(1, 0),ncol=1)
    ax_d.set_ylabel('Percentage \nof available area used')

    ax_d.set_xticklabels(['GER NUTS0', 'GER NUTS1', 'GER MAX-P'], rotation=0)
    ax_d.tick_params(length=0)
    ax_d.legend(['Onshore wind','Solar','Offshore wind'], loc='lower left', bbox_to_anchor=(1, 0.1),ncol=1)

    for side in ['top','right']:
        ax_d.spines[side].set_visible(False)

    # ---

    plt.text(-0.16, 2, '(a)',fontsize=13, transform=ax_b.transAxes, fontweight='semibold')
    plt.text(-0.16, 0.8, '(b)',fontsize=13, transform=ax_b.transAxes, fontweight='semibold')
    plt.text(-0.16, 0.8, '(c)',fontsize=13, transform=ax_c.transAxes, fontweight='semibold')
    plt.text(-0.16, 0.8, '(d)',fontsize=13, transform=ax_d.transAxes, fontweight='semibold')

    fig3.savefig('plots/case_studies_results.pdf',bbox_inches='tight')
plot_graphs(8,13)
```
